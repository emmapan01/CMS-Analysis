### Import Data

```python
# import hvpb_person dataset
import pandas as pd
df_hvbp = pd.read_csv("/content/drive/My Drive/Colab Notebooks/CIS 450/hvbp_person.csv") # Updated file path
df_hvbp
```

### Hospital Quality Ranking System

```python
# Convert 'Facility ID' column in df_hvbp to string
df_hvbp['Facility ID'] = df_hvbp['Facility ID'].astype(str)

# Merge on Facility ID
merged_df = pd.merge(df_info, df_hvbp, on="Facility ID", how="inner")

# Show the first 5 rows of the merged dataset
print(merged_df.head())

# Optionally save to CSV
merged_df.to_csv("merged_hospital_quality_data.csv", index=False)
```

```python
selected_columns = [
    "Overall Rating Of Hospital Performance Rate",
    "Hcahps Base Score",
    "Hcahps Consistency Score",
    "Communication With Nurses Floor"
]
```

```python
import numpy as np

for col in selected_columns:
    # Replace % and non-numeric values like 'Not Available'
    merged_df[col] = (
        merged_df[col]
        .str.replace('%', '', regex=False)
        .replace("Not Available", np.nan)
        .astype(float)
    )
    # Fill missing values (optional)
    merged_df[col] = merged_df[col].fillna(0)  # or use mean(), median() if preferred
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
merged_df[selected_columns] = scaler.fit_transform(merged_df[selected_columns])
```

```python
weights = {
    "Overall Rating Of Hospital Performance Rate": 0.4,
    "Hcahps Base Score": 0.2,
    "Hcahps Consistency Score": 0.2,
    "Communication With Nurses Floor": 0.2
}

# Compute weighted score
merged_df["Quality Score"] = sum(
    merged_df[col] * weight for col, weight in weights.items()
)
```

```python
merged_df["Rank"] = merged_df["Quality Score"].rank(ascending=False)
merged_df_sorted = merged_df.sort_values(by="Quality Score", ascending=False)

# Show top 10 hospitals
merged_df_sorted[["Facility Name_x", "City/Town_x", "State_x", "Quality Score", "Rank"]].head(10)
```

```python
merged_df_sorted = merged_df_sorted.reset_index(drop=True)
merged_df_sorted[["Facility Name_x", "City/Town_x", "State_x", "Quality Score", "Rank"]].head(10)
```

Merge rank + ownership to analyze how each type performs overall.

```python
rank_df = merged_df[[
    "Facility ID",
    "Facility Name_x",
    "City/Town_x",
    "State_x",
    "Hospital Ownership",  # Or rename if it’s slightly different
    "Quality Score",
    "Rank"
]]
```

```python
ownership_summary = rank_df.groupby("Hospital Ownership").agg({
    "Quality Score": "mean",
    "Rank": "mean",
    "Facility ID": "count"
}).rename(columns={
    "Quality Score": "Avg Quality Score",
    "Rank": "Avg Rank",
    "Facility ID": "Hospital Count"
}).sort_values("Avg Quality Score", ascending=False)

print(ownership_summary)
```

### Combine Ownership and ranking

```python
import matplotlib.pyplot as plt

ownership_summary["Avg Quality Score"].plot(kind="barh", figsize=(10, 6), title="Average Quality Score by Hospital Ownership")
plt.xlabel("Average Quality Score")
plt.ylabel("Hospital Ownership")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

Merge this chart with readmission data

### Combine readmission with quality score

```python
readmission_summary = df_readmissions.groupby("Facility ID").agg({
    "Excess Readmission Ratio": "mean"
}).reset_index()
```

```python
# Merge ranking and readmission ratios
merged_analysis = pd.merge(rank_df, readmission_summary, on="Facility ID")
```

```python
# Compare ownership types across both Quality Score and Readmission Ratio
ownership_comparison = merged_analysis.groupby("Hospital Ownership").agg({
    "Quality Score": "mean",
    "Excess Readmission Ratio": "mean",
    "Facility ID": "count"
}).rename(columns={
    "Quality Score": "Avg Quality Score",
    "Excess Readmission Ratio": "Avg Readmission Ratio",
    "Facility ID": "Hospital Count"
}).sort_values("Avg Quality Score", ascending=False)

print(ownership_comparison)
```

```python
ownership_comparison[["Avg Quality Score", "Avg Readmission Ratio"]].plot(
    kind='barh',
    figsize=(10, 6),
    title="Hospital Ownership: Avg Quality Score vs Readmission Ratio"
)
plt.xlabel("Score / Ratio")
plt.ylabel("Hospital Ownership")
plt.tight_layout()
plt.show()
```

Correlate nurse communication with overall hospital ratings

### Correlate nurse communication with overall hospital ratings

```python
# Filter and clean nurse communication data
df_nurse_comm = df_patient[
    df_patient["HCAHPS Question"] == 'Patients who reported that their nurses "Always" communicated well'
].copy()

# Clean percentage column
df_nurse_comm["HCAHPS Answer Percent"] = (
    df_nurse_comm["HCAHPS Answer Percent"]
    .replace("Not Available", None)
    .str.replace('%', '', regex=False)
    .astype(float)
)

# Keep only what we need
df_nurse_comm = df_nurse_comm[["Facility ID", "HCAHPS Answer Percent"]]
df_nurse_comm = df_nurse_comm.rename(columns={"HCAHPS Answer Percent": "Nurse Communication Score"})
```

```python
# Extract overall rating, remove 'Not Available'
df_ratings = df_info[["Facility ID", "Hospital overall rating"]].copy()
df_ratings = df_ratings[df_ratings["Hospital overall rating"] != "Not Available"]
df_ratings["Hospital overall rating"] = df_ratings["Hospital overall rating"].astype(float)
```

```python
# Merge on Facility ID
df_merged_corr = pd.merge(df_nurse_comm, df_ratings, on="Facility ID")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(
    data=df_merged_corr,
    x="Nurse Communication Score",
    y="Hospital overall rating",
    scatter_kws={"alpha": 0.6}
)
plt.title("Correlation Between Nurse Communication and Overall Hospital Rating")
plt.xlabel("Nurse Communication Score (%)")
plt.ylabel("Overall Hospital Rating (1–5 stars)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Drop missing values in both columns
df_corr_cleaned = df_merged_corr.dropna(subset=["Nurse Communication Score", "Hospital overall rating"])

# Then compute Pearson correlation
from scipy.stats import pearsonr
corr, p = pearsonr(df_corr_cleaned["Nurse Communication Score"], df_corr_cleaned["Hospital overall rating"])

print(f"Pearson correlation: {corr:.3f} (p = {p:.3e})")

```
